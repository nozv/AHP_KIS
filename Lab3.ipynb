Интенсивность потока подготовки программ $\lambda = \frac{k}{t_1}$, интенсивность потока выполнения программ $\mu = \frac{1}{t_2}$

$\rho = \frac{\lambda}{\mu}$

```r
l <- k/t1
m <- 1/t2
rho <- l/m
rho
```

Найдем вероятность того, что программа не будет выполнена сразу же, то есть в системе больше одной программы:

 $$P_0 = 1 - \rho $$

 $$P_1 = \rho \cdot (1 - \rho)$$

 $$P_{k>1} = 1 - P_0 - P_1 = 1 - (1 - \rho + \rho(1 - \rho)) = 1 - (1 - \rho^2) = \rho^2$$

```r
rho^2
```

Найдем среднее время до получения пользователем результатов реализации, то есть среднее время программы в системе:

$$ T_{сист} = \frac{1}{\mu(1-\rho)} $$


```r
T_sist
```

Найдем среднее количество программ, ожидающих выполнения на сервере, то есть среднюю длину очереди:

$$ L_{оч} = \frac{\rho^2}{(1-\rho)} $$

```r
L <- rho^2/(1-rho); L
```

Проведем моделирование системы, и убедимся, что экспериментальные данные совпадают с теоретическими расчетами:

```r
#install.packages("simmer")
library(simmer)
```

```r
#моделирование
SIM_TIME = 1000000

program <- trajectory("program's path") %>%
   set_global('count', 1, mod='+') %>%
   branch(function() ifelse(get_queue_count(firm, "server") > 0, 1, 0),
          continue = TRUE,
          trajectory() %>%
          set_global('nonzero',1, mod='+')
   ) %>%
   seize("server") %>%
   timeout(function() rexp(1, 1/t2)) %>%
   release("server")

distr <- function() rexp(1, 1/t1)
firm <- simmer("firm") %>%
   add_global("count",0) %>%
   add_global("nonzero",0) %>%
   add_resource("server", mon=TRUE) %>%
   add_generator(paste0("Programmer№", c(1:k)," "), program, from(distr, distr), mon=2)

firm %>% run(until = SIM_TIME)
```

Последовательность положений точки можно рассматривать как неоднородную цепь Маркова. Переходные вероятности $P_{i,i}=p_1, P_{i,i+1}=p_2, P_{i,i-1}=p_3$, остальные равны нулю. Искомая вероятность $P(k) = x_{-m}(k)+x_{-m+1}(k)+...+x_{-1}(k)+x_0(k)+x_1(k)+...+x_{m-1}(k)+x_m(k)$, где $x_i(k)$ - вероятность нахождения точки в координате $i$ на $k$-м шаге.

$x_i(k) = x_{i-1}(k-1)p_2+x_i(k-1)p_1+x_{i+1}(k-1)p_3$

Начальное положение точки $x_0(0) = 1$

На первом шаге $x_{-1}(1)=p_3, x_{0}(1)=p_1, x_1(1)=p_2$

```r
#вероятности нахождения точки на первом шаге
ax <- function(j)
{
  if (j==-1) return(p3)
  else if (j==0) return(p1)
  else if (j==1) return (p2)
  else return(0)
}

#рекурсивное вычисление вероятностей нахождения точки в i-той координате на t-м шаге
rec <- function(i, t) {
  if (t==1)
    res <- ax(i-1)*p2+ax(i)*p1+ax(i+1)*p3
  else
    res <- rec(i-1,t-1)*p2+rec(i,t-1)*p1+rec(i+1,t-1)*p3
  return(res)
}
```
