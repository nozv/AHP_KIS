Проведем моделирование:

```r
install.packages("simmer")
library(simmer)
```

```r
SIM_TIME = 300000

fail <- trajectory("failures") %>%
   timeout(function() rexp(1, nu)) %>%                                          #интервал отказов
   set_capacity("channel", 0) %>%                                               #отказ канала
   timeout(function() rexp(1, gamma)) %>%                                       #ожидание восстановления работоспособности
   set_capacity("channel",1)   %>%                                              #восстановление работоспособности
   rollback(5, Inf)                                                             #зацикливание траектории

req <- trajectory("request's path") %>%
   seize("channel") %>%                                                         #попадание программы на обработку или в очередь
   timeout(function() rexp(1, mu)) %>%                                          #время выполнения программы
   release("channel")                                                           #освобождение канала

distr <- function() rexp(1, lambda)

syst <- simmer("system") %>%
   add_resource("channel", capacity=1, queue_size = 4, queue_size_strict=4,     #канал с длиной очереди m
   mon=TRUE, preemptive = T) %>%
   add_generator("failure", fail, at(0), priority = 10) %>%                     #генерация отказов канала
   add_generator(paste0("req", c(1:k)," "), req, from(0, distr, F),
   priority=0, restart=TRUE)


syst %>% run(until = SIM_TIME) #%>% invisible
```

```r
req_stats <- get_mon_arrivals(syst, per_resource = T)
chnl_stats <- get_mon_resources(syst)
chnl_stats$delta <- c(chnl_stats[1,]$time, diff(chnl_stats$time))
```

```r
print("P_prostoy")
Exp_pr <- nrow(chnl_stats[chnl_stats$system == 0,])/nrow(chnl_stats); Exp_pr

print("P_och")
Exp_poch <- nrow(chnl_stats[chnl_stats$queue != 0 & chnl_stats$capacity !=0,])/nrow(chnl_stats[chnl_stats$capacity !=0,]); Exp_poch

print("L_och")
Exp_loch <- mean(chnl_stats$queue); Exp_loch

print("W_och")
Exp_woch <- mean(req_stats$end_time - req_stats$start_time - req_stats$activity_time); Exp_woch

print("L_sist")
Exp_lsist <- mean(chnl_stats$system); Exp_lsist

print('T_sist')
Exp_tsist <- mean(req_stats$end_time - req_stats$start_time); Exp_tsist

print('Lambda')
Exp_lam <- nrow(req_stats)/SIM_TIME; Exp_lam
```

```r
Parameter <- c("P_простоя", "P_оч", "L_оч", "W_оч", "L_сист", "T_сист", "Lambda_абс")
Theory <- c(P[1], P_och, L_och, W_och, L_sist, T_sist, lmbda)
Experiment <- c(Exp_pr, Exp_poch, Exp_loch, Exp_woch, Exp_lsist, Exp_tsist, Exp_lam)
Result <- data.frame(Parameter, Theory, Experiment)
View(Result)
```
